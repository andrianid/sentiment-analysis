# Mean Calories of Different Categories

```python
#Rata-rata Kalori di tiap Kategori
MeanCalories = {}
for category in df['Category']:
    mean = int(df[df['Category'] == category]['Calories'].mean())
    MeanCalories[category] = mean
pd.Series(MeanCalories).sort_values(ascending=True).plot.barh().set_title("Mean Calories of Different Categories")
```

```python
totCal = df['Calories'].sum()
```

```python
df1=df.groupby(['Category'],as_index=True).sum() #
df1/totCal*100
```

```python
df1['Calories']
```

## Check how much do beverages, like soda or coffee, contribute to the overall caloric intake

```python
CoffeeSoda = df1['Calories'][[1,4]].sum()/totCal
Others = df1['Calories'][[0,2,3,5,6,7,8]].sum()/totCal
cal = {'Coffee and Soda':[CoffeeSoda], 'Foods and Others':[Others]}
df2 = pd.DataFrame(data=cal).transpose()
```

```python
df2.plot(kind='pie',
            figsize=(9,12),
            autopct='%1.2f%%',
            startangle=90,
            shadow=False,
            subplots=True,
            fontsize=20
        )

plt.title('Calories Intake')
plt.axis('equal')
plt.ylabel('')

plt.show()
```

# Question 3
## Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value? 

#### Filter Item "Sandwich"

```python
df_sandwich = df[df["Item"].str.contains("Sandwich")]
```

```python
df_sandwich
```

```python
df_sandwich["Sandwich Category"] = df_sandwich["Item"].map(lambda x: "Crispy" if "Crispy" in x else "Grilled" if "Grilled" in x else "")
```

```python
df_sandwich
```

```python
df_sandwichALL = df_sandwich.groupby("Sandwich Category").mean()
```

```python
df_sandwichALL = df_sandwichALL.reset_index()
```

```python
df_sandwichALL
```

```python
df_sandwichALL["Total Nutritions"] = df_sandwichALL["Total Fat"] + df_sandwichALL["Carbohydrates"] + df_sandwichALL["Dietary Fiber"] +  df_sandwichALL["Protein"] + (df_sandwichALL["Vitamin A (% Daily Value)"]/100)*2000 + (df_sandwichALL["Vitamin C (% Daily Value)"]/100)*2000 + (df_sandwichALL["Calcium (% Daily Value)"]/100)*2000 + (df_sandwichALL["Iron (% Daily Value)"]/100)*2000
```

```python
df_sandwichALL
```

```python
ax = df_sandwichALL.plot.barh(x="Sandwich Category", y ="Total Nutritions").set_title(" Crispy Sandwich vs Grilled Sandwich")
```

```python
df_sandwich["Total Nutritions"] = df_sandwich["Total Fat"] + df_sandwich["Carbohydrates"] + df_sandwich["Dietary Fiber"] +  df_sandwich["Protein"] + (df_sandwich["Vitamin A (% Daily Value)"]/100)*2000 + (df_sandwich["Vitamin C (% Daily Value)"]/100)*2000 + (df_sandwich["Calcium (% Daily Value)"]/100)*2000 + (df_sandwich["Iron (% Daily Value)"]/100)*2000
```

```python
df_sandwich.reset_index()
```

```python
classic = df_sandwich.loc[[57,58]]
club = df_sandwich.loc[[59,60]]
ranch = df_sandwich.loc[[61,62]]
bacon = df_sandwich.loc[[63,64]]
```

```python
san1 = classic.plot.barh(x="Sandwich Category", y ="Total Nutritions").set_title("Crispy Classic Sandwich vs Grilled Classic Sandwich")
san2 = club.plot.barh(x="Sandwich Category", y ="Total Nutritions").set_title("Crispy Club Sandwich vs Grilled Club Sandwich")
san3 = ranch.plot.barh(x="Sandwich Category", y ="Total Nutritions").set_title("Crispy Ranch BLT Sandwich vs Grilled Ranch BLT Sandwich")
san4 = bacon.plot.barh(x="Sandwich Category", y ="Total Nutritions").set_title("Bacon Clubhouse Crispy Sandwich vs Bacon Clubhouse Grilled Sandwich")
```
